```python
bronxpr = df[df['municipio'] == 'Bronx']['room_type'].value_counts()[0]
bronxhome = df[df['municipio'] == 'Bronx']['room_type'].value_counts()[1]
bronxshared =df[df['municipio'] == 'Bronx']['room_type'].value_counts()[2]
brooklynpr = df[df['municipio'] == 'Brooklyn']['room_type'].value_counts()[0]
brooklynhome = df[df['municipio'] == 'Brooklyn']['room_type'].value_counts()[1]
brooklynshared =df[df['municipio'] == 'Brooklyn']['room_type'].value_counts()[2]
mpr = df[df['municipio'] == 'Manhattan']['room_type'].value_counts()[0]
mhome = df[df['municipio'] == 'Manhattan']['room_type'].value_counts()[1]
mshared =df[df['municipio'] == 'Manhattan']['room_type'].value_counts()[2]
qpr = df[df['municipio'] == 'Queens']['room_type'].value_counts()[0]
qhome = df[df['municipio'] == 'Queens']['room_type'].value_counts()[1]
qshared =df[df['municipio'] == 'Queens']['room_type'].value_counts()[2]
stpr = df[df['municipio'] == 'Staten Island']['room_type'].value_counts()[0]
sthome = df[df['municipio'] == 'Staten Island']['room_type'].value_counts()[1]
stshared =df[df['municipio'] == 'Staten Island']['room_type'].value_counts()[2]
```

```python
valores1 = [mhome,mpr,mshared]
valores2 = [bronxhome ,bronxpr ,bronxshared]
valores3 = [brooklynhome,brooklynpr,brooklynshared]
valores4 = [qhome,qpr,qshared ]
valores5 = [sthome, stpr , stshared]
```

```python
fig, ax = plt.subplots(figsize=(14,10))
et = ['Entire home', 'Private room', 'Shared room ']
x = np.arange(len(et)) 
ancho = 0.15
plt.bar(x - 0.4, valores1, ancho, color = 'Green')
plt.bar(x - 0.23, valores2, ancho, color = 'Blue')
plt.bar(x - 0.052 , valores3, ancho, color = 'Red')
plt.bar(x + 0.13, valores4, ancho, color = 'Magenta')
plt.bar(x + 2*ancho, valores5, ancho, color = 'Brown')

plt.plot([],[], color='green', label = 'Manhattan')  # solo sirve para la legend
plt.plot([],[], color='blue', label = 'Bronx')  
plt.plot([],[], color='red', label = 'Brooklyn')
plt.plot([],[], color = 'magenta', label = 'Queens')
plt.plot([],[], color = 'brown', label = 'Staten Island')
ax.set_xticks(x) 
ax.set_xticklabels(et)
plt.legend(loc=7)  
plt.title('Habitaciones por municipio')  
plt.show()
```

## 4.	Graficar la cantidad de reviews de cada municipio de acuerdo a tipo de habitación

```python
#valores1 = bronxentire, brooklyentire, mentire, queensetnire, statenentire - blue
bronxentire = df[(df['municipio'] == 'Bronx')& (df['room_type']=='Entire home/apt')]['number_of_reviews'].sum()
bronxnpr = df[(df['municipio'] == 'Bronx')& (df['room_type']=='Private room')]['number_of_reviews'].sum()
bronxshared = df[(df['municipio'] == 'Bronx')& (df['room_type']=='Shared room')]['number_of_reviews'].sum()

brooklynentire = df[(df['municipio'] == 'Brooklyn')& (df['room_type']=='Entire home/apt')]['number_of_reviews'].sum()
brooklynpr = df[(df['municipio'] == 'Brooklyn')& (df['room_type']=='Private room')]['number_of_reviews'].sum()
brooklynshared = df[(df['municipio'] == 'Brooklyn')& (df['room_type']=='Shared room')]['number_of_reviews'].sum()

mentire = df[(df['municipio'] == 'Manhattan')& (df['room_type']=='Entire home/apt')]['number_of_reviews'].sum()
mnpr = df[(df['municipio'] == 'Manhattan')& (df['room_type']=='Private room')]['number_of_reviews'].sum()
mshared = df[(df['municipio'] == 'Manhattan')& (df['room_type']=='Shared room')]['number_of_reviews'].sum()

qentire = df[(df['municipio'] == 'Queens')& (df['room_type']=='Entire home/apt')]['number_of_reviews'].sum()
qpr = df[(df['municipio'] == 'Queens')& (df['room_type']=='Private room')]['number_of_reviews'].sum()
qshared = df[(df['municipio'] == 'Queens')& (df['room_type']=='Shared room')]['number_of_reviews'].sum()

stentire = df[(df['municipio'] == 'Staten Island')& (df['room_type']=='Entire home/apt')]['number_of_reviews'].sum()
stpr = df[(df['municipio'] == 'Staten Island')& (df['room_type']=='Private room')]['number_of_reviews'].sum()
stshared = df[(df['municipio'] == 'Staten Island')& (df['room_type']=='Shared room')]['number_of_reviews'].sum()
```

```python
valores1 = [bronxentire,brooklynentire,mentire, qentire,  stentire ]
valores2 = [bronxnpr,brooklynpr,mnpr,  qpr , stpr ]
valores3 = [bronxshared,brooklynshared,mshared, qshared,  stshared ]
```

```python
fig, ax = plt.subplots(figsize=(14,10))
et = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island']
x = np.arange(len(et)) 
ancho = 0.2
plt.bar(x - 0.2, valores1, ancho, color = 'Blue')
plt.bar(x, valores2, ancho, color = 'Green')
plt.bar(x + 0.2 , valores3, ancho, color = 'Red')


plt.plot([],[], color='blue', label = 'Entire home')  # solo sirve para la legend
plt.plot([],[], color='Green', label = 'Private room')  
plt.plot([],[], color='red', label = 'Shared room')

ax.set_xticks(x) 
ax.set_xticklabels(et)
plt.legend(loc=2)  
plt.title('Reviews por municipio')  
plt.show()
```
