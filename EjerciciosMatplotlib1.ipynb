## 4. Obtener la siguiente gr√°fica

```python
fig, ax = plt.subplots(figsize=(9,6))
etiquetas=["a","b","c","d"] #eje x
valores=[10,20,30,40]
valores1=[20,30,40,50]

x = np.arange(len(etiquetas))  #0,1,2,3 
ancho=0.5

plt.bar(x - ancho/2, valores, ancho, color="aqua") # para colocarlas movida a un lado , primera barra en -0,2 
plt.bar(x + ancho/2, valores1, ancho ,color="magenta")

ax.set_xticks(x) #los puntitos sobresalientes en x
ax.set_xticklabels(etiquetas)
```
